```python
print("MINTEMP")
print("Type: ",type(dfarreglada["MinTemp"]))
print("--------------------------")
print("Promedio: ", round(dfarreglada["MinTemp"].mean(),2))

print("--------------------------")
print("Median: ",(dfarreglada["MinTemp"].median()))

print("--------------------------")
print("Mode: ",(dfarreglada["MinTemp"].mode()[0]))
```

```python
print("-------------------------- En el caso de población (si hay muchos datos):")
print("Con 0 grados de libertad porque: This dataset contains about 10 years of daily weather observations from many locations across Australia")
print("Desviación estándar de la población (con 0 grados de libertad): ",round(dfarreglada["MinTemp"].std(ddof=0),2))
print("La varianza de la población (con 0 grados de libertad): ",round(dfarreglada["MinTemp"].var(ddof=0),2))
print("--------------------------")
print("-------------------------- En el caso de muestra (si fue una parte):")
print("Con 1 grado de libertad")
print("Desviación estándar de la muestra (con 1 grado de libertad): ",round(dfarreglada["MinTemp"].std(ddof=1),2))

print("La varianza de la muestra (con 1 grado de libertad): ",round(dfarreglada["MinTemp"].var(ddof=1),2))
```

```python
fig, ax = plt.subplots(figsize = (50, 5))
sns.boxplot(data = dfarreglada, x ="MinTemp")
```

```python
Q1 = dfarreglada["MinTemp"].quantile(0.25)
Q2 = dfarreglada["MinTemp"].quantile(0.5)
Q3 = dfarreglada["MinTemp"].quantile(0.75)
maximo=  dfarreglada["MinTemp"].max()
print("Q1 =",Q1," Q3 =", Q3, "maximo = ",maximo)
IQR = Q3-Q1
print("IQR (Interquartile range ) =",IQR)
```

```python
fig, ax = plt.subplots(figsize = (50, 5))
sns.boxplot(data=dfarreglada, x="MinTemp")
```

```python
print("limite inferior ", round(Q2-(1.5*IQR),2))
print("limite supérior ", round(maximo-(1.5*IQR),2))
```

```python
fig, ax = plt.subplots(figsize = (50, 5))
sns.boxplot(data = dfarreglada, x ="MinTemp")
plt.axvline(maximo-(1.5*IQR), color="blue",label="Lim. Superior")
plt.axvline(Q3-(1.5*IQR), color="red",label="Lim. Inferior")
```

```python
sns.displot(dfarreglada["MinTemp"].values, color='darkblue', kde=True)
plt.title("MinTemp")
plt.xlabel('')
plt.ylabel('')
moda = dfarreglada["MinTemp"].mode()[0]
plt.axvline(dfarreglada["MinTemp"].mean(), color='red', label='Promedio')
plt.axvline(moda, color='red', label='Moda')
plt.axvline(dfarreglada["MinTemp"].median(), color='#4FCA28', label='Mediana')
plt.legend()
plt.show()
```
