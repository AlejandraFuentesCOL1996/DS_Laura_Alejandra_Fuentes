Sistemas No supervisados

```python

import numpy as np
import pandas as pd
import sklearn as sk
import matplotlib
import sys
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# from sklearn.model_selection import train_test_split
# from sklearn.linear_model import LogisticRegression
# from sklearn import metrics
```

```python
#Simulador de 
#Genereme un X y un Y

#número de centroides: el numero de clasificaciones
#cluster_std = la desviacion estandar
X,y = make_blobs (n_samples = 100, n_features=2, centers= 2, cluster_std = 1, random_state=10)

#X es la variable , Y es la varaible dependiente
```

```python
y
```

```python
#Vamos a graficar con 4 espacios
fig, axs = plt.subplots(1,4,figsize=(20,4)) #[1,2,3,4]

#En el primer espacio, graficaremos: Dos grupos separados 
#Me mostrará la Y original
axs[0].scatter(X[:,0],X[:,1], c=y)

#SISTEMA NO SUPERVISADO: Le estamos pidiendo que clasifique los puntos que están en X como usted peuda, según los criterios que yo le daré y prediga Y
#Buscará los cluster que previamente ha generado
k_means_cluster= KMeans(n_clusters=2, max_iter=10,
                        init="random")

#Crearemos nuestros clusters
clusters=k_means_cluster.fit_predict(X)
center = k_means_cluster.cluster_centers_
clusters

#El cluster 
axs[1].scatter(X[:,0],X[:,1], c=clusters)
axs[1].scatter(center[:,0],center[:,1])

#Los colores de la gráfica no implica nada, solo es para mostrar pedagogicamente que son diferentes clusters. 
#Lo importante es que el sistema está logrando clasificarlos 
```

```python
O,R = make_blobs (n_samples = 500, n_features=2, centers= 4, cluster_std = 4, random_state=10)

fig,axs = plt.subplots(11,1, figsize = (4,40))#[1,2,3,4]

axs[0].scatter(O[:,0],O[:,1],c = R)
for i in range(1,10):
  kmeans_cluster = KMeans(n_clusters = 4, max_iter= i,
                          init='random')
  clusters2 = kmeans_cluster.fit_predict(O)
  center2 = kmeans_cluster.cluster_centers_
  clusters2

  axs[i].scatter(O[:,0],O[:,1],c = clusters2)
  axs[i].scatter(center2[:,0],center2[:,1])
```

```python
#AHORA CON TRES CARACTERÍSTICAS (N_FEATURES)
#DOS CATEGORIAS (CENTERS=2)
V,k = make_blobs (n_samples = 500, n_features=3, centers= 2, cluster_std = 1, random_state=10)

fig,axs = plt.subplots(11,3, figsize = (12,40))#[1,2,3]

axs[0][0].scatter(V[:,0],V[:,1],c = k)
axs[0][1].scatter(V[:,1],V[:,2],c = k)
axs[0][2].scatter(V[:,0],V[:,2],c = k)
for i in range(1,10):
  kmeans_cluster = KMeans(n_clusters = 2, max_iter= i,
                          init='random')
  clusters3 = kmeans_cluster.fit_predict(V)
  center3 = kmeans_cluster.cluster_centers_
  clusters3

#Los centroides siempre tendrán el mismo numero de columnas que el 
  axs[i][0].scatter(V[:,0],V[:,1],c = clusters3)
  axs[i][0].scatter(center3[:,0],center3[:,1])
  axs[i][1].scatter(V[:,1],V[:,2],c = clusters3)
  axs[i][1].scatter(center3[:,1],center3[:,2])
  axs[i][2].scatter(V[:,0],V[:,2],c = clusters3)
  axs[i][2].scatter(center3[:,0],center3[:,2])
```
